```python
import random, time

def choose_rps():
    "output: randomly returns rock, paper, or scissors"
    r = random.randint(0,2)
    if r == 0:
        return "Rock"
    elif r == 1:
        return "Scissors"
    else:
        return "Paper"
    
def play_again():
    r = random.randint(0, 1)
    return r

    
def rps():
    '''
    input: each player could be rock, player or scissors
    output: once the code is run, the return statement should be "it's
    a tie" or "Player 1 won" or "Player 2 won"
    '''
    # Outcome of the game
    result_1 = 'It\'s a tie'
    result_2 = 'Player 2 wins'
    result_3 = 'Player 1 wins'

    # Player choice
    print("player 1 chooses: ")
    time.sleep(1)
    player_1 = choose_rps()
    print(player_1)
    time.sleep(2)
    print("player 2 chooses: ")
    player_2 = choose_rps()
    print(player_2)

    play = True
    while play:
        # When the result is a tie
        if player_1 == player_2 == 'Rock' or player_1 == player_2 == 'Paper' or\
        player_1 == player_2 == 'Scissors':
            print(f"{result_1}\n")
            print('------')
        
        # When Player 2 wins
        elif player_1 == 'Rock' and player_2 == 'Paper':
            print(f"{result_2}\n")
            print('------')

        elif player_1 == 'Paper' and player_2 == 'Scissors':
            print(f"{result_2}\n")
            print('------')

        elif player_1 == 'Scissors' and player_2 == 'Rock':
            print(f"{result_2}\n")
            print('------')

        # When Player 1 wins
        else:
            print(f"{result_3}\n")
            print('----------')
        
        play = play_again()
        # Player choice
        print("player 1 chooses: ")
        time.sleep(1)
        player_1 = choose_rps()
        print(player_1)
        time.sleep(2)
        print("player 2 chooses: ")
        player_2 = choose_rps()
        print(player_2)

rps()

print('++++++++++++++++++++++')
print("Thank you for playing!")
print("Game Over!")

# complete the program here
```

```python
import random

def choose_rps():
    '''
    Function allows to randomly choose player choice
    of rock, paper or scissors
    '''
    r = random.randint(0, 2)  # retuns num 0 - 2 randomly

    if r == 0:
        return 'Rock'
    elif r == 1:
        return 'Scissors'
    else:
        return 'Paper'

def rps(player_1, player_2):
    '''
    input: the player could be rock, paper or scissors.
    output: it returns
    '''
    if player_1 == player_2 == 'Rock' or player_1 == player_2 == 'Paper' or\
        player_1 == player_2 == 'Scissors':
        print(f"It's a tie!")
        print('------')
    
    elif player_1 == 'Paper' and player_2 == 'Rock' or player_1 == 'Scissors'\
    and player_2 == 'Paper' or player_1 == 'Rock' and\
    player_2 == 'Scissors':
        print('Player 1 wins!')

    else:
        print('Player 2 wins!')

def play_again():
    r = random.randint(0, 1)
    
play = True

print("Welcome to Rock, Paper, Scissors!\n")
while play == True:
  player1 = choose_rps()
  player2 = choose_rps()

  print("Player 1 chose", player1, ".")
  print("Player 2 chose", player2, ".")

  rps(player1, player2)

  print(" ")

  play = play_again()

print("Thank you for playing!")



```
