```python
import random

names_std = []
shortened_name = []
id_std = []
email_std = []
std_index = []

def enter_name_again():
    response = input("Do you want to enter another details? y for yes, n for no: Your response: ")
    if response == "y":
        r = 1
    else:
        r = 0
    # r = random.randint(0, 1)
    return r

def std_name():
    '''
    This function adds students full name into the
    names_std list
    '''

    name = []
    first_name = input('Enter your first name? ')
    last_name = input('Enter your last name: ')
    full_name = first_name + ' ' + last_name
    names_std.append(full_name)

    if ' ' in first_name:
        name = first_name.split()
        name = name[0][0] + name[1][0] + last_name
    else:
        name = first_name[0] + last_name
    shortened_name.append(name)
    return shortened_name


def std_id():
    '''
    This function generates random numbers 
    between 11111 and 99999 as students id
    '''
    
    for name, index in enumerate(names_std):
        number = random.randint(111111, 999999)
        if len(names_std) > len(id_std) and number not in id_std:
            std_index.append(number)
            id_std.append(number % 1000)

    return id_std

def account_gen():
    '''
    Here, the account_gen function calls the
    other two functions to generate the students email
    '''
    enter_name = True
    while enter_name:
        std_name()
        std_id()
        enter_name = enter_name_again()



    for index, name in enumerate(shortened_name):
        email_std.append(name + str(id_std[index]) + '@adadev.org')

    for index, name in enumerate(names_std):
        print(f"name: {name}")
        print(f"id:{std_index[index]}")
        print(f"email:{email_std[index]}")
        print('\n')


# std_name()
# std_id()
account_gen()


```
